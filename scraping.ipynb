```python
def parse(url):
    response = requests.get(url)
    tree = html.fromstring(response.text)
    
    title_path = '/html/body/div[1]/div[1]/div/main/div/div[1]/div/div/h1/span'
    title = tree.xpath(title_path)[0].text_content().strip()
    
    subtitle_path = '/html/body/div[1]/div[1]/div/main/div/div[2]/div[1]/p/strong'
    subtitle = tree.xpath(subtitle_path)[0].text_content().strip()
    
    text_path = '/html/body/div[1]/div[1]/div/main/div/div[2]/div[2]'
    selected_div = tree.xpath(text_path)[0]
    elements_to_exclude = selected_div.xpath('.//aside')
    for div_element in selected_div.xpath('.//div'):
            div_element.getparent().remove(div_element)
    for nested_element in nested_elements_to_exclude:
        nested_element.getparent().remove(nested_element)
    div_direct_text_content = ' '.join([i.strip() for i in selected_div.xpath('.//text()')])     
    text = div_direct_text_content.strip()
    text = text.split('Читайте також:')[0].strip().replace('  ',' ')
    
    return {"url": url, "title":title, "subtitle": subtitle, "text": text}
```

```python
def find_articles(url):
    response = requests.get(url)
    tree = html.fromstring(response.text)
    
    articles = []
    urls_to_scan = [url]
    for i in range(2,101):
        urls_to_scan.append(f'{url}/page-{i}')
    for article_url in urls_to_scan:
        response = requests.get(article_url)
        tree = html.fromstring(response.text)
        for i in range(1, 10):
            try:
                x_path = f'/html/body/div[1]/div/div/main/div[3]/div/div[{i}]/article/div/div/h3/a'
                selected_element = tree.xpath(x_path)[0]
                element_text = selected_element.get('href')
                articles.append(element_text)
            except:
                pass
    return articles
```

```python
def parse_all(articles):
    parsed = []
    for url in articles:
        try:
            parsed.append(parse(url))
        except:
            pass
    return parsed
```

```python
def parse(url):
    response = requests.get(url)
    tree = html.fromstring(response.text)
    
    title_path = '/html/body/div[1]/div[1]/div/main/div/div[1]/div/div/h1/span'
    title = tree.xpath(title_path)[0].text_content().strip()
    
    subtitle_path = '/html/body/div[1]/div[1]/div/main/div/div[2]/div[1]/p/strong'
    subtitle = tree.xpath(subtitle_path)[0].text_content().strip()
    
    text_path = '/html/body/div[1]/div[1]/div/main/div/div[2]/div[2]'
    selected_div = tree.xpath(text_path)[0]
    elements_to_exclude = selected_div.xpath('.//aside')
    for div_element in selected_div.xpath('.//div'):
            div_element.getparent().remove(div_element)
    for nested_element in nested_elements_to_exclude:
        nested_element.getparent().remove(nested_element)
    div_direct_text_content = ' '.join([i.strip() for i in selected_div.xpath('.//text()')])     
    text = div_direct_text_content.strip()
    text = text.split('Читайте також:')[0].strip().replace('  ',' ')
    
    return {"url": url, "title":title, "subtitle": subtitle, "text": text}
```

```python
def find_articles(url):
    response = requests.get(url)
    tree = html.fromstring(response.text)
    
    articles = []
    urls_to_scan = [url]
    for i in range(2,3001):
        urls_to_scan.append(f'{url}?page-{i}')
    for article_url in urls_to_scan:
        response = requests.get(article_url)
        tree = html.fromstring(response.text)
        for i in range(1, 21):
            try:
                x_path = f'/html/body/div[1]/div/div/main/div[3]/div/div[{i}]/article/div/div/h3/a'
                selected_element = tree.xpath(x_path)[0]
                element_text = selected_element.get('href')
                articles.append(element_text)
            except:
                pass
    return articles
```

```python
def parse_all(articles):
    parsed = []
    for url in articles:
        try:
            parsed.append(parse(url))
        except:
            pass
    return parsed
```
